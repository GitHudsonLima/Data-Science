````python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('Profissão- Cientista de Dados_M10_support material.csv') 
#df.head()
#df.info()
#df.shape
#df[['data_ref']].value_counts()
````

````python
df1 = df[['renda','posse_de_veiculo','posse_de_imovel','data_ref']]
df1.head()
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda", data=df1,  height=4, ci = 90 )
````

##### A renda é mais alta para quem tem posse de veículo

````python
ax = sns.pointplot(x="data_ref", y="renda", hue='posse_de_veiculo', data=df1,  height=4, ci = 90 )
legend = plt.legend(loc = 'lower center',bbox_to_anchor = (0.5, -0.5),ncol =3)
ticks = plt.xticks(rotation = 90)
title = plt.title('Posse de veiculo por renda e data referência')
````

#### Analisando pela data de referência do dado verificamos também que quem tem posse de veículo também possui uma renda maior.

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda", data=df1,  height=4, ci = 90 )
````

#### A renda de quem tem a posse de imóvel aparentemente é maior.

````python
ax = sns.pointplot(x="data_ref", y="renda", hue='posse_de_imovel', data=df1,  height=4, ci = 90 )
legend = plt.legend(loc = 'lower center',bbox_to_anchor = (0.5, -0.5),ncol =3)
ticks = plt.xticks(rotation = 90)
title = plt.title('Posse de Imóvel por renda e data referência')
````

#### Analisándo o gráfico junto as datas de referência verificamos que:
* Nem sempre a renda de quem tem posse de imóvel é maior
* A diferença de renda entre quem tem posse de imóvel e quem não tem posse de imóvel não é tão representativa quando comparado a posse de veículos
