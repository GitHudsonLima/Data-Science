## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui
#(operações com escalar) Calcule o percentual de acerto dos alunos por assunto.
df_mat['%_acertos'] = (df_mat/24)*100 
df_por['%_acertos'] = (df_por/18)*100 
df_geo['%_acertos'] = (df_geo/8)*100
df_ing['%_acertos'] = (df_ing/8)*100
df_his['%_acertos'] = (df_his/8)*100
df_fis['%_acertos'] = (df_fis/12)*100
df_qui['%_acertos'] = (df_qui/12)*100

````

````python
# 2) Seu código aqui
#(operações entre *DataFrames) Calcule o total de acertos de cada aluno.
df_total = df_mat[['Qt_acertos']] + df_por[['Qt_acertos']] + df_geo[['Qt_acertos']] + df_ing[['Qt_acertos']] + df_his[['Qt_acertos']] + df_fis[['Qt_acertos']] + df_qui[['Qt_acertos']]
df_total
#aprovacoes.loc[:,'acre':]
````

````python
# 3) Seu código aqui
#Calcule o porcentual geral de cada aluno.
df_total['%_acertos_total'] = (df_total['Qt_acertos']/90)*100 
df_total
````

````python
# 4) Seu código aqui
#Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?
df_total_bool = (df_total['Qt_acertos']>45)
Qtd_alunos_notalM45 = df_total_bool.sum()
print(f'Quantidade de alunos com nota maior que 45: \n {Qtd_alunos_notalM45}\n')
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# seu código aqui
# 1) Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. 
#Monte um novo DataFrame com os alunos desqualificados por este critério.
df_mat_zero = df_mat[df_mat['Qt_acertos'] == 0]
df_fis_zero = df_fis[ df_fis['Qt_acertos']==0]
df_qui_zero = df_qui[ df_qui['Qt_acertos'] ==0]

#lista dos desqualificados
list_desqualificado = [df_mat_zero, df_fis_zero, df_qui_zero]

#concatenação dos desqualificados
df_desqualificado = pd.concat(list_desqualificado)

#print(f'df_desqualificado :\n {df_desqualificado} ')

#retirando as duplicidades de index - alunos que zeraram em mais de uma materia (mat, fis ou qui)
df_desqualificado_dup = df_desqualificado[~df_desqualificado.index.duplicated(keep='first')]

#ordenando os index
df_desqualificado_order = df_df_desqualificado_dup.sort_index()

#print(f'df_desqualificado.sort_index() :\n {df_desqualificado.sort_index()} \n')
#print(f'df_desqualificado.drop_duplicates() :\n {df_desqualificado.drop_duplicates()}\n ')
print(f'Alunos desqualificados ordenados e sem duplicidade :\n {df_desqualificado_order} \n')

````

````python
#verificando tempo dos metodos para retirar duplicidade dos index
#%timeit df_desqualificado[~df_desqualificado.index.duplicated(keep='first')]
#%timeit df_desqualificado.groupby(df_desqualificado.index).first()
````

````python
# 2) Quantos são esses alunos?
Qtd_alunos_desqualificados = len(df_desqualificado_order.index)
print(f'Qtd_alunos_desqualificados \n {Qtd_alunos_desqualificados} \n')
````

````python
# 3) Médias dos alunos em história e geografia
    #If you’d like to select rows based on integer indexing, you can use the .iloc function.
    #If you’d like to select rows based on label indexing, you can use the .loc function.

#lista dos alunos desqualificados
lt_desqualificados = df_desqualificado_order[df_desqualificado_order['Qt_acertos']==0].index.tolist()
#print(lt_desqualificados)

#Media desses alunos em Geografia
df_geo_desqualificados = df_geo.loc[lt_desqualificados]
#print(f'df_geo_desqualificados \n {df_geo_desqualificados}\n')
mean_df_geo_desqualificados = df_geo_desqualificados['Qt_acertos'].mean()
print(f'Media desses alunos em Geografica \n {mean_df_geo_desqualificados} \n')

#Media desses alunos em História
df_his_desqualificados = df_his.loc[lt_desqualificados]
#print(f'df_his_desqualificados \n {df_his_desqualificados}\n')
mean_df_his_desqualificados = df_his_desqualificados['Qt_acertos'].mean()
print(f'Media desses alunos em Historia \n {mean_df_his_desqualificados} \n')

#Media desses alunos em História e Geografia
lt_geo_his_desqualificados = [df_geo_desqualificados, df_his_desqualificados]
dt_geo_his_desqualificados = pd.concat(lt_geo_his_desqualificados)
#print(f'dt_geo_his_desqualificados \n {dt_geo_his_desqualificados}')

mean_df_geo_his_desqualificados = dt_geo_his_desqualificados['Qt_acertos'].mean()
print(f'Media desses alunos em Historia e Geografia juntando os dois dataframes \n {mean_df_geo_his_desqualificados} \n')
````
